```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Trajectories.csv", delimiter=",")
# some attributes are outside scope, dropped
df = df.iloc[:, :4]
df.columns = ['id', 'frame', 'x', 'y']
df
```

```python
# visualizing
plt.figure(figsize=(20, 10))
plt.scatter(df.x, df.y, s=0.5, c=df.id)
```

```python
# we can already tell something's wrong: id=2750 starts at frame 664, we assumed every particle was present in the
# molecule since frame 0 and that their number was constant. Is this true? 
# How many molecules do we have at any given time?
by_frame = df.groupby(by='frame')
by_frame['id'].describe()
```

```python
# So we start with 82 molecules, we go up to 118, and by the end only half of them are present.
# Therefore, resonably many IDs are subsequent parts of their trajectories which were wrongly classified as new molecules.
# We need to find their original trajectories and stich as many as possible to maximise their predictability.
# First attempt will be finding pairs of molecules where the last frame of one precedes the first frame of another.
# any time we find a single pair (meaning, there are only two IDs with this characteristic in that particular frame,
# we have a high chance they are the same molecule).
```

```python
# First step is to get for every frame the current IDs.
frames = []
tmp = by_frame['id'].apply(pd.DataFrame).transpose()
for i, row in tmp.iterrows():
    frames.append(row.dropna().tolist())
print(frames[:2]) # To show IDs are removed and added
```

```python
# With this we want to find out for every frame which IDs are removed and added.
previous_frame = frames[0]
frame_number = 0
for current_frame in frames[1:]:
    frame_number += 1
    dead_ids = []
    new_ids = []
    
    # dead IDs list
    for id in previous_frame:
        if id not in current_frame:
            dead_ids.append(id)
    
    # new IDs list
    for id in current_frame:
        if id not in previous_frame:
            new_ids.append(id)

    # Position check: if a new id is very close to a dead one, match them.
    for dead_id in dead_ids:
        potential_matches = df[['x', 'y']].loc[df.id == dead_id].loc[df.frame == frame_number]
    for new_id in new_ids:
        potential_matched = df[['x', 'y']].loc[df.id == new_id].loc[df.frame == frame_number]
    # Closeness measure (in pixels).
    c = 30
    # If the dead and the new are closer than c in both x and y dimensions, they become the same trajectory.
    if not potential_matches.empty and not potential_matched.empty:
        for dead in potential_matches:
            for new in potential_matched:
                if dead.x - c <= new.x and new.x <= dead.x + c and dead.y - c <= new.y and new.y <= dead.y + c:
                    print(dead.id, 'becomes', new.id)
    
    previous_frame = current_frame
```
